```python
def transColName(colname):

    if(translate[translate["EN"]==colname]["TW"].count()>0):
        return translate[translate["EN"]==colname]["TW"].values[0]
    else:
        return colname
```

```python
#  categorical 
categorical = [var for var in x_train.columns if x_train[var].dtype=='O']
print('There are {} categorical variables'.format(len(categorical)))
```

```python
#  numerical 
numerical = [var for var in x_train.columns if x_train[var].dtype!='O']
print('There are {} numerical variables'.format(len(numerical)))
```

```python
sns.distplot(train['SalePrice']);
```

```python
i = 0
threshold = 20
discrete = []
dataDisplay = pd.DataFrame(columns=['E-column', 'C-column','Unique values'])
for var in numerical:
    if len(x_train[var].unique())<threshold:
        discrete.append(var)
        dataDisplay.loc[i] = [var, transColName(var), x_train[var].unique()]
        i += 1

       
print('There are {} discrete variables'.format(len(discrete)))

#HTML(tabulate(dataDisplay, numalign='right', headers=['E-column', 'C-column', 'Unique values'], tablefmt='html'))
```

```python
continuous = [var for var in numerical if var not in discrete and var not in ['Id', 'SalePrice']]
```

```python
listColumns = []

listValues = []

i = 0

dataDisplay = pd.DataFrame(columns=['E-column', 'C-column', 'Total', 'Null', 'Percent'])

for var in x_train.columns:

    total = x_train[var].isnull().count()

    countNum = x_train[var].isnull().sum()

   

    if countNum>0 and var != 'SalePrice':

        listColumns.append(transColName(var))

        listValues.append(round(x_train[var].isnull().mean(),2))

        nullPercent = round(x_train[var].isnull().mean() * 100,1)

        dataDisplay.loc[i] = [var, transColName(var), total, countNum, nullPercent]

        i += 1

#HTML(tabulate(dataDisplay, numalign=right, tablefmt=html))
```

```python
data = pd.DataFrame({'x': listColumns, 'y': listValues})

Chart(data).mark_bar().encode(x='y', y='x')
```
